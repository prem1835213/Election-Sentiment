```python
def bag_of_words_features(vocab, document):
    """
    vocab is a list of word stems
    document is a list of word stems
    """
    features = np.zeros(len(vocab))
    for i in range(len(vocab)):
        features[i] = document.count(vocab[i])
    return features
```
